### Aquí vamos a realizar un análisis de frecuencia para identificar los valores más comunes o frecuentes en cada clúster

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import QuantileTransformer, OneHotEncoder
from sklearn.compose import ColumnTransformer
import pandas as pd
```

```python
pd.pandas.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('data_complete.csv')
```

```python
data.head(30)
```

```python
data_sinCluster = data[data.columns.difference(['Cluster'])]
```

```python
categorical_features = ['causa_def', 'edad_agru', 'ent_ocules', 'ent_ocurr', 'ent_resid', 'gr_lismex', 'lista1', 'lista_mex', 'loc_ocurr', 'loc_resid', 'lugar_ocur', 'mun_ocurr', 'mun_resid']

numerical_features = data_sinCluster.columns.difference(categorical_features)
numerical_features = numerical_features.tolist()
```

```python
data_sinCluster[categorical_features]
```

```python
data_sinCluster[numerical_features]
```

```python
# Preprocessing pipeline
numeric_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median', add_indicator=True)),
    ('encoding', QuantileTransformer(output_distribution='normal')),
    # ('encoding', StandardScaler())

])

categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='None', add_indicator=True)),
    ('encoding', OneHotEncoder(max_categories=50))
    # ('encoding', OneHotEncoder())
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_pipeline, numerical_features),
        ('cat', categorical_pipeline, categorical_features)
    ])

preprocessor
```

Use

```python
preprocessed_features = preprocessor.fit_transform(data_sinCluster)
preprocessed_features = preprocessed_features.toarray()
```

```python
data_sinCluster
```

```python
data_preprocesada = pd.DataFrame(preprocessed_features,
            #  columns=data.columns,
             index=data.index)
```

```python
data_preprocesada = pd.concat([data_preprocesada, data[['Cluster']]],
                    axis=1)
```

```python
data_preprocesada
```

### Modeling

```python
from pycaret.classification import *
```

```python
# Configurar el entorno de PyCaret
setup(data=data, target='Cluster', numeric_features=numerical_features, categorical_features=categorical_features, session_id=123)
```

```python
# Obtener los datos transformados después de la etapa de preprocesamiento
X_train_transformed = get_config('X_train')
X_train_transformed
```

```python
# Obtener los datos transformados después de la etapa de preprocesamiento
X_train_transformed = get_config('X_train_transformed')
X_train_transformed
```
